## <font color=green> Example 1:
### Program to Check If a Given Year Is a Leap Year. 
nestedifexample.png



```python
year=int(input("enter a year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f'{year} is a leap year')
        else: print(f'{year} is not a leap year')
    else:
        print(f'{year} is a leap year')
else:
    print(f'{year} is not a leap year')
```
