# <font color=red> **pandas**
- pandas is a Python library that provides fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both easy and intuitive.
- pandas has two primary data structures:
  - Series (one-dimensional) and 
  - DataFrame (two-dimensional)
  
- pandas is built on top of NumPy
- pandas is well suited for:
  - inserting and deleting columns from DataFrame, 
  - easy handling of missing data (represented as NaN), 
  - explicitly aligning data to a set of labels,
  - converting data in other Python and NumPy data structures into DataFrame objects,
  - intelligent label-based slicing, indexing, and subsetting of large data sets, 
  - merging and joining of data sets, and flexible reshaping. 
  - Additionally, it has robust input/output tools for loading data from CSV files, Excel files, databases, and other formats. 
  - You have to import a pandas library to make use of various functions and data structures defined in pandas.

- pandas is usually renamed as pd

# DataFrame Indexing and Selecting Data: loc[] & iloc[]


- .loc[] and .iloc[] are used to retrieve rows.
- They are called as indexers
- The .loc[] method is primarily label based
- The .loc[] method will raise KeyError when the items are not found


- The .iloc[] method is primarily an integer position based (from 0 to length-1 of the axis)
- The .iloc[] method will raise an IndexError if a requested indexer is out-of-bounds, except in the case of slice indexers, which allow out-of bounds indexing.

```python
import pandas as pd
pd.read_excel('reg_list.xlsx').head()
```

```python
reg_df=pd.read_excel('reg_list.xlsx', usecols=[0, 1, 2, 6, 7, 8, 9, 10])
reg_df.head()
```

# split it into xlsx files such that each file contains the data of just one subject


```python
reg_gp=reg_df.groupby(['كود المادة'])

import os
for course_code, course_data in reg_gp:
  file_name=course_code+'.xlsx'
  course_data.to_excel(file_name, sheet_name='sheet 1', index=False)
```

# print the course list that every student register

```python
st_course_list=reg_df.groupby('كود الطالب')
for st_code, st_data in st_course_list:
  print(st_code)
  st_data=st_data.reset_index(drop=True)
  print(st_data['كود المادة'])
print(st_course_list)
```

# Count the number of succeeded students in each subject

```python
all_courses=reg_df.groupby(['كود المادة'])
for course_code, course_data in all_courses:
  print(course_code)
  #print(course_data)
  print(course_data[course_data['المجموع']>=60]['المجموع'].count())    #print(sum(course_data['المجموع']>=60))
  
```

# Compute the precentage of Excellent grade for each subject


```python
all_courses=reg_df.groupby(['كود المادة'])
for course_code, course_data in all_courses:
  print(course_code)
  print(sum(course_data['المجموع']>=90)*100 / course_data.shape[0])
  
```
