#<font color=red> Numpy
- NumPy is the fundamental package for scientific computing with Python. 

- It stands for “Numerical Python.” 

- It supports interacting with massive multidimensional array structures. 

- Learning NumPy is the first step in every beginner-level data scientist's path
 
   - because it is the base module for nearly all of the powerful libraries of Python's data science toolkit (including Pandas, Matplotlib, and Scikit-learn). 


- NumPy’s main object is the homogeneous multidimensional array. 

- An array is a table of elements (usually numbers), all of the same type, indexed by a tuple of positive integers and represented by a single variable. 

- NumPy’s array class is called ndarray. It is also known by the alias array.

## <font color=blue> read numpy files using genfromtxt()

- The genfromtxt() function is used quite frequently to load data from text files in python. 

- We can read data from CSV files using this function and store it into a numpy array. 

- This function has many arguments available, making it a lot easier to load the data in the desired format. 

- We can specify the delimiter, deal with missing values, delete specified characters, and specify the datatype of data using the different arguments of this function.
