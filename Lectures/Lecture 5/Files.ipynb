# change the position of the file handler using <font color=blue>seek(position)

```python
file_name=input("enter the filename: ")
with open(file_name, mode='w+', encoding='UTF-8') as file:
   file.write("hey world!")
   print(file.tell())
   print(file.seek(0))
   print(file.tell())
   print(file.read())
   print(file.tell())
```

#**<font color=red>Exercise 2:**

Write Python Program to Reverse Each Word in "hey.txt" file.


```python
with open("hey.txt", 'r', encoding='UTF-8') as file:
  lines=file.readlines()
reversed_lines=''
for line in lines:
  words=line.strip().split(" ")
  for word in words:
     reversed_lines+=word[::-1]
     reversed_lines+=' '
  reversed_lines+='\n'
print(reversed_lines) 
```

# <font color=red>Exercise 3</font>

 Write Python Program to Count the Occurrences of Each Word and Also Count the Number of Words in a "hey.txt" File.

```python
with open ("hey.txt", 'r', encoding='utf-8') as file:
  lines=file.readlines()
no_of_words=0
words_count={}
for line in lines:
      words=line.strip().split(" ")
      no_of_words+=len(words)
      for word in words:
        words_count[word]= words_count[word]+1  if word in words_count   else 1
        
print(words_count)
print("the total number of words in the file is: {}".format(no_of_words))
```

```python
with open ("hey.txt", 'r', encoding='utf=-8') as file:
  lines=file.readlines()

longest_word=''
for line in lines:
  line_words=line.strip().split(' ')
  for word in line_words:
    if len(word) >len(longest_word): longest_word= word
print("the longest word in the file is: {}. with size: {}".format(longest_word, len(longest_word)))
```

# <font color=red>Exercise 2:

Write Python program to demonstrate the writing of data to a CSV file using DictWriter class

```python
with open('names.csv', 'w', newline='') as csvfile:
  field_names = ['first_name', 'last_name']
  writer = csv.DictWriter(csvfile, fieldnames=field_names)
  writer.writeheader()
  writer.writerow({'first_name': 'Baked', 'last_name': 'Beans'})
  writer.writerow({'first_name': 'Lovely', 'last_name': 'Spam'})
  writer.writerow({'first_name': 'Wonderful', 'last_name': 'Spam'})
  writer.writerows([{'first_name': 'W', 'last_name': 'S'}, {'first_name': 'e', 'last_name': 'r'}, {'first_name': 'u', 'last_name': 't'}])
```
